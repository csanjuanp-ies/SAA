## Qué es el machine Learning
- **Inteligencia artificial**: Todas las técnicas que intentan simular la inteligencia humana en una computadora
- **ML**: Conjunto de algortimos matemáticos que intentan mejorar las tareas con la experiencia extrayendo
conocimiento de los datos a través de algoritmos matemáticos.
- **Deep Learning**. Conjunto de técnicas que permiten mejorar ciertas tareas avanzadas mediante redes neuronales
y gran cantidad de datos.

*“Machine learning is the field of study that gives computers the ability to learn without being explicitly programmed”
 Arthur L. Samuel, AI pioneer, 1959*

**El ML es el proceso de obtención de conocimiento a partir de los datos**. EL ML ha provocado un cambio en el
paradigma de la programación.

## Tipos de inteligencia Artificial
Según el tipo de problema que solucionan
- **Artificial Intelligence (AI)**: Campo de la ciencia de los computadores que se encarga de resolver tareas humanas
de manera eficiente (natural language, speech, image recognition, ...)
- **Artificial General Intelligence (AGI)**: IA multipropósito que intenta imitar la inteligencia humana.
- **Narrow AI**: IA que resuleve una tarea concreta (playing a game, driving a car, ...)

Las IA se clasifican en IA Débiles o Fuertes
- **Débil**: Tareas concretas, no aprende fuera del ámbito de su problema, sin flexibilidad, programada, no razonan.
- **Fuerte**. Todo lo contrario, no existen.

## Conceptos del ML
### Tipos de variables en el ML
- **Cuantitativas**. Valores numéricos: edad, ingresos, altura.
    - **Continuas**
    - **Discreteas**
- **Ordinales**. Expresan un orden: ranking
- **Cualitativas**. Expresan una cualidad: color, sexo.

### Tipos de aprendizaje en el ML
- **Supervisado**: Se parte de datos etiquetados ya conocidos con los que entrenaremos el modelo.
Buscaremos predicciones actuales y futuras o clasificación de los datos de entrada. Las etiquetas deben ser
discretas (contables). Las variables se suelen llamar características y la variable de salida es la variable objetivo.
    - Por ejemplo, **clasificar** correo spam o determinar la clase de una imagen.
    - O intentaremos realizar una **predicción o regresión**, donde a partir de un conjunto de variables se consigue una
predicción.
    - En un último tipo de aprendizaje se ordenan un conjunto de objetos de entrada según alguna característica,
es el llamado **regresión ordinal**.
- **No supervisado**. Partimos de datos sin etiquetas para buscar una estructura interna, sin una función de recompensa y sin datos etiquetados.
El ejemplo principal es el clustering o agrupación sin conocer nada más que los propios datos. Cada clúster tendrá características similares.
Por ejemplo, se podrá determinar grupos de compradores basados en sus intereses.
Otra aplicación es la disminución de la dimensionalidad, en la que se reducen el número de variables de entrada sin
pérdida en la información subyacente, este proceso puede eliminar gran parte del ruido.
- **Reforzado**. Es el proceso de decisión mediante un sistema de recompensas para aprender una serie de acciones.
El objetivo es hacer un sistema que mejore su rendimiento basado en las interacciones con el entorno.
El feedback se realiza a través de medidas que nos dicen cómo de buena es la acción a través de una función de
recompensa. El agente aprende a través de la interacción con el entorno y un proceso de prueba y error.
Un ejemplo típico es un motor de ajedrez en el que agente tratará de maximizar la función de recompensa
(ganar la partida) a través de interacciones con el entorno (jugadas). Un concepto importante es que la recompensa
se puede obtener de forma inmediata o al final (como es el caso del ajedrez).

### Clasificación vs Regresión
Un modelo de **regresión** predice valores continuos. Por ejemplo, los modelos de regresión hacen predicciones que responden a preguntas como las siguientes:
¿Cuál es el valor de una casa en California?
¿Cuál es la probabilidad de que un usuario haga clic en este anuncio?

![](img/ut01_03.png)

Un modelo de **clasificación** predice valores discretos. Por ejemplo, los modelos de clasificación hacen predicciones que responden a preguntas como las siguientes:
¿Un mensaje de correo electrónico determinado es spam o no es spam?
¿Esta imagen es de un perro, un gato o un hámster?

## Desafíos del ML
- **Datos insuficientes**. La captura de datos es primordial para cualquier proyecto de ML, no solo la cantidad debe ser
adecuada, sino también la calidad de los mismos. En general serán necesarios miles de ejemplos para poder hacer
converger a nuestros modelos. Si bien el que no haya suficientes ejemplos en un hándicap, hay algoritmos que podrán
sacar partido de ellos aun siendo insuficientes.
- **Datos no representativos**. Es imprescindible que los datos sean representativos de los casos que queremos representar.
- **Baja calidad de los datos**. Si los datos son erróneos o contienen ruido el algoritmo no podrá converger de forma
eficiente. Por lo que limpiar los datos y adecuarlos es una tarea primordial antes de todo.
- **Características irrelevantes**. Nuestro sistema solo será capaz de aprender adecuadamente si los datos son correctos,
y si hay características no necesarias deben ser eliminadas antes de empezar el proceso.
- **Sobreajuste de los datos**. El sobreajuste se produce cuando un modelo no generaliza bien para datos no presentados
en el entrenamiento, este es un grave problema que deberemos tratar.
- **Subajuste de los datos**. En este caso el modelo es demasiado simple para recoger todas las características de los datos.

## Pasos en el ML

![](img/ut01_09.png)

- **Preprocesado**. Es una de las etapas más importantes en la que vamos a preparar los datos para los algoritmos.
De esta fase dependerá el éxito de las siguientes fases. Muchos algoritmos requieren que las características estén
en la misma escala o sean comparables. O puede que haya datos redundantes y se tenga que eliminar.
En esta fase dividiremos los datos en dos conjuntos, uno de entrenamiento y otro de pruebas.
- **Selección y entrenamiento**. Elegiremos el modelo que más se ajusta a los datos, siendo muy importante comparar
el resultado de varios algoritmos para encontrar el mejor. Para poder comparar el resultado es imprescindible elegir
una manera de medir el rendimiento entre todos los algoritmos. Hay que ajustar los parámetros por defecto del modelo.
- **Evaluación**. La única manera de saber si un modelo generaliza bien con datos no utilizados es evaluando el modelo.
Para ello utilizaremos los datos de pruebas para estimar el error de generalización.
Si el error de generalización sobre los datos de prueba es muy bajo, pero sobre los datos de test es alto indicará que
nuestro modelo se ha sobreajustado
- **Predicción**. Si la evaluación es aceptable usaremos el modelo para pasar a las predicciones o clasificaciones reales
